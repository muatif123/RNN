#### Essentially to predict for the next 24 hours, we dont actually need the 3 months data. Instead we just use the last 1 month data for training the model

```python
df = df['2016-05-01':]
```

```python
df = df.round(2)
```

```python
df
```

```python
len(df)
```

```python
test_days = 2
```

```python
test_index = test_days * 144      # 24*60/10
```

```python
test_index
```

```python
train = df.iloc[:-test_index]
test = df.iloc[-test_index:]
```

```python
train
```

```python
test
```

#### Scaling the Data

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
```

```python
scaled_train = scaler.fit_transform(train)
scaled_test = scaler.transform(test)
```

```python
from tensorflow.keras.preprocessing.sequence import TimeseriesGenerator
```

```python
# Defining the generator
length = 144

generator = TimeseriesGenerator(scaled_train, scaled_train, length = length, batch_size = 1)
```

```python
len(scaled_train)
```

```python
len(generator)
```

```python
X, Y  = generator[0]
```

#### Model Building

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, LSTM
```

```python
scaled_train.shape
```

```python
model = Sequential()

model.add(LSTM(25, input_shape = (length, scaled_train.shape[1])))

model.add(Dense(scaled_train.shape[1]))

model.compile(optimizer = 'adam', loss = 'mse')
```

```python
from tensorflow.keras.callbacks import EarlyStopping
early_stop = EarlyStopping(monitor = 'val_loss', patience = 2)
```

```python
validation_generator = TimeseriesGenerator(scaled_test, scaled_test, length = length, batch_size = 1)
```

```python
%%time
model.fit_generator(generator, epochs = 30, validation_data = validation_generator, callbacks = [early_stop])
```

```python
losses = pd.DataFrame(model.history.history)
```

```python
losses.plot()
```

```python
first_eval_batch = scaled_train[-length:]
```

```python
first_eval_batch = first_eval_batch.reshape((1, length, scaled_train.shape[1]))
```

```python
model.predict(first_eval_batch)
```

```python
scaled_test[0]
```

```python
n_features = scaled_train.shape[1]
test_pred = []

first_eval_batch = scaled_train[-length:]
current_batch = first_eval_batch.reshape((1, length, n_features))

for i in range(len(test)):
    current_pred = model.predict(current_batch)[0]
    test_pred.append(current_pred)
    
    current_batch = np.append(current_batch[:, 1:, :], [[current_pred]], axis = 1)
```
